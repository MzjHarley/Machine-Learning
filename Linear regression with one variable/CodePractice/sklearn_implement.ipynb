```python
from sklearn import linear_model,metrics
import numpy as np
import pandas as pd
data = pd.read_csv("../Data/data1.txt",header=None,names=['Population','Profit'])
x=np.array(data.iloc[:,:-1])
y=np.array(data['Profit'])
clf = linear_model.LinearRegression().fit(x,y)
k = clf.coef_ #parameter
b = clf.intercept_ 
y0=clf.predict(x)
print("Fitted Curve Parameters:",b,k[0])
print("cost:",metrics.mean_squared_error(y, y0)/2)
x1=np.array([[3.5]])
print("predict:",clf.predict(x1))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6),dpi=80)
plt.xlabel("Population")
plt.ylabel("Profit")
plt.scatter(data['Population'],data['Profit'],marker='o',label='Traning Data')
plt.plot(data['Population'],b + (k[0] * data['Population']),'r',label='Prediction')
plt.legend()
plt.show()
```
